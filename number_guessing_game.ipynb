```python
import random
r=random.randint(1,50)
attempts=0
maxattempts=3
print("Welcome to number guessing Game!")
print("A Secret number has been selected")
print("You have 3 attempts to guess it.")
print("----------------------------------------")
while attempts<maxattempts:
    guess=int(input("enter your guess between 1 to 50:"))
    attempts+=1
    if guess<1 or guess >50:
        print("enter a number between 1 and 50")
        attempts-=1
        continue
    elif guess==r:
        print("you guessed the number right in",attempts,"attempts")
        break
    elif guess>r:
        if attempts<maxattempts:
            print("Too high.Try again!")
        else:
            print("Too high")
    elif guess<r:
        if attempts<maxattempts:
            print("Too low.Try again!")
        else:
            print("Too low")
if guess!=r and attempts>=3:
    print("GAME OVER!The number was",r)


        
```
